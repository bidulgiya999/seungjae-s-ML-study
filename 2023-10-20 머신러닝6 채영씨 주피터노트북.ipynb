%d : 2자리 수 date (1~9일의 경우, 앞에 0을 채운다)
%H : 2자리 수 시간 (24-hour clock, 0~9시의 경우, 앞에 0을 채운다)
%M : 2자리 수 분 (0~9분의 경우, 앞에 0을 채운다)
%S : 2자리 수 초 (0~9초의 경우, 앞에 0을 채운다)

```python
def newSeason(month):
    if month in [12,1,2]:
        return 4
    elif  month in [3,4,5]:
        return 1
    elif  month in [6,7,8]:
        return 2
    elif  month in [9,10,11]:
        return 3
```

```python
train['season']=train.month.apply(newSeason)
```
