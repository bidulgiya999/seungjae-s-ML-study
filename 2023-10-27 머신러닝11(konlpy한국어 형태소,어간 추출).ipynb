```python
from math import log 
# df("바나나") = 2
# tf*idf : 문서에서 각 단어의 중요도를 나타내는 행렬
# 각 문서에서 중요단어가 몇개인지 알고자 함 => 키워드 => 토픽 모델링
# tf-idf는 단어의 중요도
# 중요하다는 조건이 필요
# tf가 크다는 건 단어가 중요하다는 것, tf가 커질수록 idf가, idf가 커질수록 tf가 커짐

"""
N = 문서의 갯수

N
_____   * tf = tfidf

df+1

"""


docs = [
  '먹고 싶은 사과',
  '먹고 싶은 바나나',
  '길고 노란 바나나 바나나',
  '저는 과일이 좋아요'
]
```

```python
vocab=list(set(w for doc in docs for w in doc.split()))
vocab
```

### 코퍼스(말뭉치) : 전체 문서에 등장하는 단어 집합

- 총 9개로 되어있는 말뭉치

```python
import pandas as pd
import numpy as np
```

```python
N = len(docs) 
def tf(t, d):
  return d.count(t)

def idf(t):
  df = 0
  for doc in docs:
    df += t in doc
  return log(N/(df+1))

def tfidf(t, d):
  return tf(t,d)* idf(t)
```

```python
result = []

# 각 문서에 대해서 아래 연산을 반복
for i in range(N):
  result.append([])
  d = docs[i] # 먹고싶은사과 부터 차례로 들어감
  for j in range(len(vocab)):
    t = vocab[j] # 말뭉치를 처음부터 끝까지("과일이") ~~~
    result[-1].append(tf(t, d))

tf_ = pd.DataFrame(result, columns = vocab)
```

```python
# 그래서 먹고싶은사과에 과일이 몇개 들어가는가 이런식으로 단어의 빈도수를 보여줌
# result에 등장횟수가 들어감
```

```python
import re
text = "I was wondering if anyone out there could enlighten me on this car."

# 길이가 1~2인 단어들을 정규 표현식을 이용하여 삭제
shortword = re.compile(r'\W*\b\w{1,2}\b') # \W ,\w{1,2} : 문자와 숫자가 최소 한들자 이상 두글자 까지  제거하겠다는 것
                                         # \b\w{1,2}\b는 앞뒤로 공백문자가 있고 그 사이에 단어가 1이상 2까지 제거한다는 것
print(shortword.sub('', text))        #shortword에 부합되는 단어 삭제
```

```python
example = "Family is not an important thing. It's everything."
stop_words = set(stopwords.words('english')) 

word_tokens = word_tokenize(example)

result = []
for word in word_tokens: 
    if word not in stop_words: 
        result.append(word) 

print('불용어 제거 전 :',word_tokens) 
print('불용어 제거 후 :',result)
```
